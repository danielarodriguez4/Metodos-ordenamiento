##Manejo de datos

```python
import pandas as pd
import math

#Leer el archivo csv
bd = pd.read_csv('Backpacks.csv')

#Eliminar columnas irrelevantes del dataset
bd = bd.drop(columns= ['name', 'main_category', 'sub_category', 'image', 'link', 'ratings', 'discount_price', 'no_of_ratings'])

#Limpiar la columna quitando las comas, dado que indica unidades de miles de los precios
bd['actual_price'] = bd['actual_price'].str.replace('₹', '').str.replace(',', '')

#Convertir los valores de la columna a float
bd = [float(x) for x in bd['actual_price']]

#Eliminar todos los datos "nan"
bd = [x for x in bd if not math.isnan(x)]
bd
```

##Algoritmo Radix Sort

```python
import time

inicio = time.time()

#Dado que este algoritmo funciona con números enteros, este método está implementado para convertir los precios con decimales a números enteros y mostrar los precios organizados con decimales al final.

def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n  # Array de salida
    count = [0] * 10  # Array de conteo para dígitos del 0 al 9

    # Contar ocurrencias de los dígitos
    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1

    # Cambiar count[i] para que contenga la posición real de los dígitos
    for i in range(8, -1, -1):  # Cambiamos el rango para contar en orden inverso y se organicen de forma descendente
        count[i] += count[i + 1]

    # Construir el array de salida en orden descendente
    for i in range(n - 1, -1, -1):
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1

    # Copiar el array de salida al array original
    for i in range(n):
        arr[i] = output[i]

def radix_sort(arr):
    # Obtener el valor máximo para determinar el número de dígitos
    max_val = max(arr)

    # Aplicar counting sort para cada dígito
    exp = 1
    while max_val // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

# Asegúrate de que esta sea la lista de precios
arr = bd

# Escalar los precios a enteros multiplicando por 100
arr_int = [int(price * 100) for price in arr]

print("Lista original (escalada):", arr_int)

# Ordenar
radix_sort(arr_int)

# Volver a escalar a precios decimales
arr_sorted = [price / 100 for price in arr_int]
print("\n")
print("Lista ordenada de forma descendente:", arr_sorted)  # Ahora ya está ordenada en descendente
print("\n")
fin = time.time()
print('Tiempo de ejecución: ', fin - inicio)
print("\n")
print("Cantidad de datos ordenados: ", len(bd))
```
